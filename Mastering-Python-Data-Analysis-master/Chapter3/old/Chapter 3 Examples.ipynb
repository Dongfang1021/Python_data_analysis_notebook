```python
def plot_cdf(data, plot_range=None, scale_to=None, **kwargs):
    sorted_data = np.array(sorted(data), dtype=np.float64)
    data_range = sorted_data[-1] - sorted_data[0]
    if plot_range is None:
        xmin = sorted_data[0]
        xmax = sorted_data[-1]
    else:
        xmin, xmax = plot_range
    count = np.arange(1, len(data) + 1)
    xvalues = np.concatenate(zip(sorted_data, sorted_data))
    xvalues = np.concatenate([[xmin], xvalues, [xmax]])
    yvalues = np.concatenate(np.array(
       [(k,k) for k in range(len(sorted_data)+1)],
       dtype=np.float64))
    if scale_to is not None:
        yvalues = yvalues/len(data) * scale_to
    return plt.plot(xvalues, yvalues, **kwargs)
```

```python
nvalues = 20
norm_variates = rnd.randn(nvalues)
plot_cdf(norm_variates, plot_range=[-3,3], scale_to=1.0, 
         lw=2.5, color='Brown')
for v in [0.25, 0.5, 0.75]:
    plt.axhline(v, lw=1, ls='--', color='black')
plt.axis([-3,3,0,1])
None
```

```python
N = 4857
mean = 63.8
serror = 0.06
sdev = serror * np.sqrt(N)
rvnorm = st.norm(loc=mean, scale=sdev)
```

```python
xmin = mean-3*sdev
xmax = mean+3*sdev
xx = np.linspace(xmin,xmax,200)
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.plot(xx, rvnorm.cdf(xx))
plt.title('Cumulative distribution function')
plt.xlabel('Height (in)')
plt.ylabel('Proportion of women')
plt.axis([xmin, xmax, 0.0, 1.0])
plt.subplot(1,2,2)
plt.plot(xx, rvnorm.pdf(xx))
plt.title('Probability density function')
plt.xlabel('Height (in)')
plt.axis([xmin, xmax, 0.0, 0.1])
None
```

```python
mean=0.0
sdev=1.0
rvnorm = st.norm(loc=mean, scale=sdev)
cdf = rvnorm.cdf
pdf = rvnorm.pdf
a = -.6
b = 1.2
xmin = mean-3*sdev
xmax = mean+3*sdev
xx = np.linspace(xmin,xmax,200)
plt.figure(figsize=(14,5))
yy = cdf(xx)
plt.subplot(1,2,1)
plt.title('Cumulative distribution function')
#plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off') 
#plt.tick_params(axis='y', which='both', left='off', right='off', labelleft='off') 
plt.plot([a,a], [0.0, cdf(a)], lw=1.2, ls='--', color='black')
plt.plot([b,b], [0.0, cdf(b)], lw=1.2, ls='--', color='black')
plt.plot([xmin,a], [cdf(a), cdf(a)], lw=1.2, ls='--', color='black')
plt.plot([xmin,b], [cdf(b), cdf(b)], lw=1.2, ls='--', color='black')
plt.plot(xx, yy, color='Brown', lw=2.5)
plt.plot([xmin, xmin], [cdf(a)+0.01, cdf(b)-0.015], lw=10, color='black')
plt.text(a, -0.06, '$a$', fontsize=18, horizontalalignment='center')
plt.text(b, -0.06, '$b$', fontsize=18, horizontalalignment='center')
plt.text(xmin-0.06, cdf(a), '$F(a)$', fontsize=18, 
         horizontalalignment='right', verticalalignment='center')
plt.text(xmin-0.06, cdf(b), '$F(b)$', fontsize=18, 
         horizontalalignment='right', verticalalignment='center')
plt.text(b, -0.06, '$b$', fontsize=18, horizontalalignment='center')
plt.text(xmin+0.15, 0.5*(cdf(a)+cdf(b)), '$P=F(b)-F(a)$', fontsize=18,
         verticalalignment='center')
yy = pdf(xx)
plt.subplot(1,2,2)
plt.title('Probability density function')
#plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off') 
#plt.tick_params(axis='y', which='both', left='off', right='off', labelleft='off') 
plt.plot([a, a], [0.0, pdf(a)], lw=1.2, color='black')
plt.plot([b, b], [0.0, pdf(b)], lw=1.2, color='black')
plt.fill_between(xx, yy, where=(a<=xx) & (xx<=b), color='LemonChiffon')
plt.text(a, -0.025, '$a$', fontsize=18, horizontalalignment='center')
plt.text(b, -0.025, '$b$', fontsize=18, horizontalalignment='center')
plt.text(0.5*(a+b)-.1, 0.2, '$P=$Area', fontsize=18,
         horizontalalignment='center',
         verticalalignment='top')
plt.plot(xx, yy, color='Brown', lw=2.5)
None
```

```python
N = 20
p = 0.5
rv_binom = st.binom(N, p)
```

```python
rv_binom.pmf(12)
```

```python
rv_binom.cdf(7)
```

```python
xx = np.arange(N+1)
cdf = rv_binom.cdf(xx)
pmf = rv_binom.pmf(xx)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
xvalues = np.concatenate(np.array(
    [(k,k) for k in range(N+1)],
    dtype=np.float64))
yvalues = np.concatenate(zip(cdf[:-1], cdf[:-1]))
yvalues = np.concatenate([[0],yvalues,[cdf[-1]]])
plt.plot(xvalues, yvalues, lw=1.5, color='brown')
plt.title('Binomial cdf, $N=20$, $p=0.5$', fontsize=16)
plt.subplot(1,2,2)
left = xx - 0.5
plt.bar(left, pmf, 1.0, color='CornflowerBlue')
plt.title('Binomial pmf, $N=20$, $p=0.5$', fontsize=16)
plt.axis([0,20,0,.18])
None
```
